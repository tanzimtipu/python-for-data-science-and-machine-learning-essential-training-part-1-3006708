```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data = {'names': ['Steve', 'John', 'Richard', 'Sarah', 'Randy', 'Michael', 'Julie'],
        'Age': [20, 22, 20, 21, 24, 23, 22],
        'Gender': ['Male' ,'Male' ,'Male' ,'Female', 'Male' ,'Male' ,'Female'],
        'Rank': [2, 1, 4, 5, 3, 7, 6]}
df = DataFrame(data)
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['names'], df['Age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Ages of people')
plt.show()
```

```python
plot = sns.barplot(x='names', y='Age', data=df)
plot.set_title('Ages of people')
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['names'], df['Age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Ages of people')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(x='names', y='Age', data=df)
plot.set_title('Ages of people')
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['Age'], labels=df['names'])
plt.title('Ages of people')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['Age'], labels=df['names'], colors=colors)
plt.title('Ages of people')
plt.show()
```
